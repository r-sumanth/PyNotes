```python
the below example can be done simply by df['price_per_person']=df['total_bill']/df['size'], 
just to get an idea of applying custom functions to multiple columns, doing this way.
```

```python
def price_per_person(price,size):
    return round((price/size),2)
```

```python
df['price_per_person']=df[['total_bill','size']].apply(lambda df: price_per_person(df['total_bill'],df['size']),axis=1)
df.head()
```

```python
np.vectorize() can be used to improve the performance of functions not already dealing with np array,
so it doesnt improve performance of functions already dealing with np array
```

```python
def discount_next(bill,tip,day):
    if day not in ['Sat','Sun'] and (tip/bill)>=0.15:
        return 10
    elif day in ['Sat','Sun'] and (tip/bill)>=0.25:
        return 15
    else:
        return 0
```

```python
if we try to apply custom functions on multiple columns without lambda
```

```python
df['next_discount_%']=df[['total_bill','tip','day']].apply(discount_next(df['total_bill'],df['tip'],df['day']),axis=1)
df.head()
```

```python
df['next_discount_%']=df[['total_bill','tip','day']].apply(lambda df: discount_next(df['total_bill'],df['tip'],df['day']),axis=1)
df.head()
```

```python
df.drop('next_discount_%',inplace=True,axis=1)
df.head()
```

```python
df['next_discount_%']=np.vectorize(discount_next)(df['total_bill'],df['tip'],df['day'])
df.head()
```

```python
df.drop('next_discount_%',axis=1,inplace=True)
df.head()
```

```python
import timeit
```

```python
setup = '''
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/sumanth/Documents/ML_PyNotes/Resource_Notebooks/03-Pandas/tips.csv')

def discount_next(bill,tip,day):
    if day not in ['Sat','Sun'] and (tip/bill)>=0.15:
        return 10
    elif day in ['Sat','Sun'] and (tip/bill)>=0.25:
        return 15
    else:
        return 0
'''
```

```python
stmnt_1 = '''
df['next_discount_%']=df[['total_bill','tip','day']].apply(lambda df: discount_next(df['total_bill'],df['tip'],df['day']),axis=1)
'''
```

```python
stmnt_2 = '''
df['next_discount_%']=np.vectorize(discount_next)(df['total_bill'],df['tip'],df['day'])
'''
```

```python
timeit.timeit(stmt=stmnt_1,setup=setup,number=10000)
```

```python
timeit.timeit(stmt=stmnt_2,setup=setup,number=10000)
```
