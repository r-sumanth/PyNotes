```python
df=pd.read_csv('/Users/sumanth/Documents/ML_PyNotes/Resource_Notebooks/05-Seaborn/dm_office_sales.csv')
df.head()
```

Scatterplot : Used for plotting points having continuous values x,y
    parameters:
        size : defines the size of points, takes in corresponding values for the points plotted 
                or even the same value given for x or y
        style : takes in categorical values, and segregates points based on marker style
        hue : colours the markers based on categorical or continuous values given 
        palette : takes in predefined (matplotlib colormaps) colour code and colours the marker 
        alpha : defines the transperancy of markers

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience',
                style='level of education',hue='level of education',palette='YlGnBu', alpha=0.75)
plt.legend(loc=(1.05,0.1))
plt.show()
```

Rugplot : Basic version of histogram, takes in only x axis and should be continuous values
    parameter :
        height : defines the height of the bars/lines in terms of percentage

```python
plt.figure(figsize=(6,2))
sns.rugplot(x='salary',data=df,height=0.35)
plt.ylim(0,1)
```

```python
# plotting categorical values to rugplot does not derive any meaning
plt.figure(figsize=(6,2))
sns.rugplot(x='level of education',data=df,height=0.35)
plt.ylim(0,1)
```

Histogram : Plot to vizualise the data distribution, takes in only for x axis, creates defined number of bins,
            y axis represents the number of data points in that bin and should be continous values
    parameters : 
        bins : defining number of bins to be created out of the values
        multiple = 'stack' : works while hue parameter is used, segregates the categorical counts on the bar
        stat : plot probability, percentage, density, frequency etc on y axis

```python
plt.figure(figsize=(5,5))
sns.set(style='darkgrid')
sns.histplot(x='salary', data=df, bins=15, hue='level of education', multiple='stack',
             palette=['orange','green','blue','purple','black'])
plt.legend(df['level of education'],loc=(1.05,0.7))
plt.show()
```

```python
sns.histplot(x='salary',data=df,stat='probability')
```

displot can ablo be used to plot a histogram, but a more genrealized version of histplot, 
as we can plot all three of them together

```python
plt.figure(figsize=(5,5))
sns.set(style='darkgrid')
sns.displot(x='salary', data=df, bins=15, 
            kde=True,rug=True,
            hue='level of education', multiple='stack',
            palette=['orange','green','blue','purple','black'])
plt.show()
```

KDE plot : Kernel Density Estimation plot, plots the probability distrbution for the dataset
            It is a method for estimating probability density distribution of random variable
            Represents the probability (y) for occurance of x
    parameters :
            clip : To get rid of values that are not in dataset range
            bw_adjust : (bandwidth) To increase the accuracy of the prabability
                        lesser value will pick up more noise because of narrow bandwidth
                        larger value will result in inaccurate plot

```python
sns.kdeplot(x='salary',data=df)
```

```python
sns.histplot(data=df['salary'],kde=True)
```

```python
np.random.seed(42)
age = np.random.randint(20,50,10)
```

```python
sns.kdeplot(age)
```

```python
sns.kdeplot(age,clip=(20,50))
```

```python
sns.kdeplot(age,clip=(20,50),bw_adjust=0.5)
```

Countplot : Plots the categorical values according to its counts
            Can use hue to further separate segregate categories
        use ax.bar_label(ax.containers[0]) to get the bar value unless using hue,
                if using hue, use for label in ax.containers:
                                      ax.bar_label(label)

```python
ax = sns.countplot(x=df['level of education'])
ax.bar_label(ax.containers[0]) #color='white',label_type='center')
plt.xticks(rotation=45)
plt.show()
```

```python
ax = sns.countplot(x=df['level of education'],hue=df['training level'].sort_values().astype('str'),
                  palette='tab10')
for label in ax.containers:
    ax.bar_label(label)
plt.xticks(rotation=45)
plt.show()
```

```python
BarPlot : Used for plotting categorical values with any measure or estimator on y axis
            Use the parameter estimator by passing required function
            use errorbar=None to get rid of black line on top of the bars
                or else sd for standard deviation, ci for confidence interval etc
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=df,x='division',y='salary',estimator='mean',errorbar='sd')
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=df,x='level of education', y='salary',estimator='mean', hue='division',errorbar='sd')
for label in ax.containers:
    ax.bar_label(label,rotation=90,label_type='center',color='white')
plt.legend(loc=(1.05,0.5))
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=df,x='division',y='sales',
                 estimator=lambda x: np.sum(x)*100.0/df['sales'].sum(),
                 errorbar=None)
ax.bar_label(ax.containers[0], fmt='%.1f',label_type='center' ,color='white')
plt.show()
```
