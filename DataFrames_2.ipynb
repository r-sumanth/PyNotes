### Dropping missing vaules

```python
df.dropna()
#this drops rows with any missing values
```

```python
# introducing the thresh parameter, 
# we can retain the columns that contain alteast a threshold number of values in that row
# here as the Jackman row contains 4 values, it wont get dropped as threshold value is set to 3
df.dropna(thresh=3)
```

```python
#now setting the threshhold value to 5, we see Jackman row gets dropped as it has only 4 values
df.dropna(thresh=5)
```

```python
df[['first_name','age','sex','post_movie_score']].dropna(thresh=3)
```

```python
# now setting the axis to 1, all the columns gets dropped 
# as there are atleast 1 instance in the column with a null value
df.dropna(axis=1)
```

```python
df.dropna(axis=0,ignore_index=True)
```

```python
# using the subset parameter,
# we can look and drop missing values in those particular columns 
df.dropna(subset=['last_name'])
```

```python
df.dropna(subset=['last_name','post_movie_score'])
```

```python
df= pd.read_csv('/Users/sumanth/Documents/ML_PyNotes/Resource_Notebooks/03-Pandas/mpg.csv')
df.head()
```

```python
# use aggregate methods like sum,avg,max,min along with group by
df.select_dtypes(include=np.number).groupby('model_year').mean()
```

```python
df.groupby('model_year').mean(numeric_only=True)
```

```python
df.groupby('model_year').mean(numeric_only=True)['mpg']
```

```python
df_grouped = df.groupby(['model_year','cylinders']).mean(numeric_only=True)
df_grouped
```

```python
df_grouped.index
```

```python
# if data of only 1 outer index grabbed, then it excludes the outer index from output
df_grouped.loc[70]
```

```python
# using multiple outer index, we can retrieve the outer index as well
df_grouped.loc[[70,73,75]]
```

```python
df_grouped.loc[70,4]
```

```python
df_grouped.loc[[(70,4),(73,4),(75,4)]]
```

```python
df_grouped.loc[[(70,4),(73,4),(75,4)],'mpg']
```

```python
df_grouped.iloc[[0,4,6]]
```

```python
df_grouped.index[[0,4,6]]
```

```python
for grabbing particular rows like retrieve all the 4 cylinder row data for all the year, 
we can use .xs function, which returns a cross section

parameter key takes in the value of rows we want to retrieve like 4 from cylinders, 
this parameter takes in only 1 value

and parameter level takes in the name of the column where the key value is located
```

```python
df_grouped.xs(key=4,level='cylinders')
```

```python
df_grouped.xs(key=(4,6),level='cylinders')
```

```python
# .xc() can also be used for outer level as well
df_grouped.xs(key=78,level='model_year')
```

```python
to retrive the result with multiple key values, we could so this by filtering out the dataset before gouping
```

```python
df[df['cylinders'].isin([4,6])].groupby(['model_year','cylinders']).mean(numeric_only=True)
```

```python
df_grouped.sort_index(level='model_year',ascending=False)
```

```python
# when inner level is sorted, it breaks the grouping

df_grouped.sort_index(level='cylinders',ascending=False)
```

```python
df_grouped.sort_values(['mpg','acceleration'],ascending=False)
```

```python
names = ['SR', 'AK', 'SN', 'VP', 'PJB', 'RRR', 'AM']
random.shuffle(names)
df1 = pd.DataFrame({'Class 8':np.random.randint(1,10,5),'Name':names[:5]})
random.shuffle(names)
df2 = pd.DataFrame({'Class 9':np.random.randint(1,10,5),'Name':names[:5]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='inner',on='Name')
```

```python
pd.merge(df1,df2,how='left',on='Name')
```

```python
pd.merge(df1,df2,how='right',on='Name')
```

```python
pd.merge(df1,df2,how='outer',on='Name')
```

```python
For columns with different names, we can use left_on,right_on,left_index,right_index parameters in combination
```

```python
df1.set_index('Name',inplace=True)
df1
```

```python
df2
```

```python
pd.merge(df1,df2,left_index=True,right_on='Name')
```

```python
df1.reset_index(inplace=True)
```

```python
pd.merge(df1,df2,left_on='Class 8',right_on='Class 9')
```
