```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(-8,8)
y = x**3
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(-8,7)
plt.ylim(-8**3,7**3)
plt.title('Cubic function')
plt.savefig('cubic_plot.png')
plt.show()
```

```python
a = np.arange(1,11)
b = a**5
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)

axes2 = fig.add_axes([0,0.7,0.5,0.5])
axes2.plot(a,b)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.set_xlabel('x axes1')
axes1.set_xlim(-8,7)
axes1.set_title('axes1')
axes1.set_ylabel('y axes1')
axes1.set_ylim(-8**3,7**3)
axes1.plot(x,y)

axes2 = fig.add_axes([0.6,0,0.4,0.5])
axes2.set_xlabel('x axes2')
axes2.set_xlim(0,10)
axes2.set_title('axes2')
axes2.set_ylabel('y axes2')
axes2.set_ylim(0,10**5)
axes2.plot(a,b)
```

```python
#figsize parameter inside the plt.figure call will size the figure in inches which will be huge when exported
#dpi dots/inch will maintain the resolution of image
fig = plt.figure(figsize=(5,3),dpi=250)

axes1 = fig.add_axes([0,0,1,1])
axes1.set_xlabel('x axes1')
axes1.set_xlim(-8,7)
axes1.set_title('axes1')
axes1.set_ylabel('y axes1')
axes1.set_ylim(-8**3,7**3)
axes1.plot(x,y)

axes2 = fig.add_axes([0.6,0,0.4,0.5])
axes2.set_xlabel('x axes2')
axes2.set_xlim(0,10)
axes2.set_title('axes2')
axes2.set_ylabel('y axes2')
axes2.set_ylim(0,10**5)
axes2.plot(a,b)

fig.savefig('plot_wo_bbox.png')
```

```python
fig = plt.figure(figsize=(5,3),dpi=250)

axes1 = fig.add_axes([0,0,1,1])
axes1.set_xlabel('x axes1')
axes1.set_xlim(-8,7)
axes1.set_title('axes1')
axes1.set_ylabel('y axes1')
axes1.set_ylim(-8**3,7**3)
axes1.plot(x,y)

axes2 = fig.add_axes([0.6,0,0.4,0.5])
axes2.set_xlabel('x axes2')
axes2.set_xlim(0,10)
axes2.set_title('axes2')
axes2.set_ylabel('y axes2')
axes2.set_ylim(0,10**5)
axes2.plot(a,b)

#with the bbox parameter will shall retain the x and y ticks and labels
fig.savefig('plot_w_bbox.png',bbox_inches='tight')
```

```python
Use plt.tight_layout() to have the plots adjusted rightly in terms of spacing
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)

axes[1,1].plot(a,b)

plt.tight_layout()
```

```python
if needed to manually adjust the space between plots, use the plt.subplots_adjust()
2 usually use parameters : wspace - Width b/w pltots in terms of ratio of width of plots
                           hspace - Height b/w pltots in terms of ratio of width of plots
Using these parameters, size of the plots will decrease
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

plt.subplots_adjust(wspace=0.3)

axes[0,0].plot(x,y)

axes[1,1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

plt.subplots_adjust(wspace=1)

axes[0,0].plot(x,y)

axes[1,1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

plt.subplots_adjust(wspace=0.5,hspace=0.3)

axes[0,0].plot(x,y)

axes[1,1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,figsize=(10,8))

axes[0].plot(x,y)
axes[0].set_title('Power 3 plot')

axes[1].plot(a,b)
axes[1].set_title('Power 5 plot')

fig.suptitle('Power plot 3v5',size=20)

plt.tight_layout()

plt.savefig('SubPlots.png',bbox_inches='tight')
```

```python
for setting a custom linestyling, use .set_dashed([]) for the plot oblect created
passing in a list indicating the patter we define, [solid_length,blank_length,...] in a recursive way
```

```python
line = plt.plot(x,y)
line[0].set_dashes([5,2,10,3,13,8])
# 5 solid length, followed by 2 blank length, 10 solid, 3 blank, 13 solid, 8 blank and repeat..
```

```python
plt.plot(x,y)[0].set_dashes([8,4,18,3])
```
